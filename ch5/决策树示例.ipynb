### 相亲决策模型

```python
from sklearn import tree
```

```python
# 训练样本集
blind_date_X = [ [35, 176, 0, 20000],
                 [28, 178, 1, 10000],
                 [26, 172, 0, 25000],
                 [29, 173, 2, 20000],
                 [28, 174, 0, 15000] ]
blind_date_y = [ 0, 1, 0, 1, 1 ]
# 测试样本集
test_sample = [  [24, 178, 2, 17000],
                 [27, 176, 0, 25000],
                 [27, 176, 0, 10000]  ]
```

```python
clf = tree.DecisionTreeClassifier() # 实例化
clf = clf.fit(blind_date_X, blind_date_y) # 建树
```

```python
clf.predict(test_sample) # 预测
```

```python
tree.plot_tree(clf) # 画出树结构
```

```python
print(clf.feature_importances_) # 给出特征的重要度
```
